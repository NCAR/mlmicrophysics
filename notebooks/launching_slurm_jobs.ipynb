```python
import yaml
```

```python
l2_weight = [1.0e-3, 1.0e-4, 1.0e-5]
lrs = [0.001, 0.0001, 0.00001]
hidden_layers = [2, 3, 4]
activation = ['relu', 'tanh']
i = 0
```

```python
def get_text_sbatch(i):
    text_sbatch = f"""#!/bin/bash -l
#SBATCH -J casp_nb
#SBATCH --account=NAML0001
#SBATCH -t 8:00:00
#SBATCH --mem=128G
#SBATCH -n 1
#SBATCH --gres=gpu:v100:1
#SBATCH -o casp_nb_{i}.o
#SBATCH -e casp_nb_{i}.e
module load gnu/8.3.0 openmpi/3.1.4 python/3.7.5 cuda/10.1
ncar_pylib ncar_20200417
export PATH="/glade/work/ggantos/ncar_20200417/bin:$PATH"
    
pip install /glade/u/home/ggantos/mlmicrophysics/.
python ../../scripts/train_mp_neural_nets.py cesm_tau_run5_full_train_nn_{i}.yml"""

    return text_sbatch 
```

```python
y = yaml.load(open('../config/cesm_tau_run5_full_train_nn_new.yml'), Loader=yaml.FullLoader)
```

```python
for act in activation:
    y['classifier_networks']['activation'] = act
    y['regressor_networks']['activation'] = act
    for hl in hidden_layers:
        y['classifier_networks']['hidden_layers'] = hl
        y['regressor_networks']['hidden_layers'] = hl
        for lr in lrs:
            y['classifier_networks']['lr'] = lr
            y['regressor_networks']['lr'] = lr
            for l2 in l2_weight:
                y['classifier_networks']['l2_weight'] = l2
                y['regressor_networks']['l2_weight'] = l2
                y['out_path'] = f'/glade/p/cisl/aiml/ggantos/200607/cam_run5_models_{i}/'
                dump_path = f'../config/200607/cesm_tau_run5_full_train_nn_{i}.yml'
                with open(dump_path, 'w') as f:
                    yaml.dump(y, f)
                dump_path = f'../config/200607/batch_{i}.sh'      
                with open(dump_path, 'w') as f:
                    f.write(get_text_sbatch(i))
                i += 1
```
